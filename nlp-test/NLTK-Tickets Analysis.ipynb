```python
def removestop(string):
    string=re.sub('[^A-Za-z0-9]+', ' ', string)
    string=string.lower()
    word_tokens = word_tokenize(string) 
    filtered_sentence = []
    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)
    return filtered_sentence    
def entit(x):
    li=[]
    for i in x:
        li.append(i)
    return ' '.join(li)
```

```python
dp=pd.DataFrame(columns=['Chat_History'])
dp['Chat_History']=grouped_df.apply(lambda x:entit(x['Chat_History']))
```

<h2>POS ---->

```python
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
```

```python
dp['POS']=dp['Chat_History'].apply(lambda x: preprocess(x))
```

```python
def find_POS(d):
    dir={}
    for i in d:
        if i[1] not in dir_.keys():
            dir_.update({i[1]:[]})
            dir_[i[1]].append(i[0])
        else:
            dir_[i[1]].append(i[0])
    return dir_
```

```python
dp['POS']=dp['POS'].apply(lambda x: find_POS(x))
```
