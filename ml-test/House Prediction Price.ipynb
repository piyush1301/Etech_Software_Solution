<h2> Missing value Treatment

```python
df=copy.copy(df_) # Backup 1
df.Garden.fillna('nan',inplace=True)
df.Sorcerer.fillna('nan',inplace=True)
df.renovation.fillna('nan',inplace=True)
df.tree_stands.fillna('nan',inplace=True)
df.Dock.fillna('Distance from the Dock is -1 holy lights',inplace=True)
df.Capital.fillna('Distance from Capital is -1 holy lights',inplace=True)
df.Royal_Market.fillna('Distance from Royal Market is -1 holy lights',inplace=True)
df.Guarding_Tower.fillna('Distance from Guarding Tower is -1 holy lights',inplace=True)
df.dining_rooms.fillna('There are -1 dining rooms',inplace=True)
df.bathrooms.fillna('There are -1 bathrooms\n',inplace=True)
df.bedroom.fillna('There are -1 bedrooms\n',inplace=True)
df.drop('visit_house',axis=1,inplace=True)
df.Location.fillna('Location of the house is : nan',inplace=True)
df.Knight_house.fillna('Distance from Knight\'s house is -1 holy lights',inplace=True)                   
```

<h2> Feature Engineering 

```python
df['House_id']=df['House_id'].apply(lambda x:re.findall(r'House ID :  (.*?)$',x)[0])
df['Date_Built']=df['Date_Built'].apply(lambda x:re.findall(r'Date Built : (.*?)$',x)[0].replace('and Date Priced :','').replace('and Date Priced','')) 
df['Capital']=df['Capital'].apply(lambda x:re.findall(r'Distance from Capital is (.*?) holy lights',str(x))[0])
df['Royal_Market']=df['Royal_Market'].apply(lambda x:re.findall(r'Distance from Royal Market is (.*?) holy lights',str(x))[0])
df['Dock']=df['Dock'].apply(lambda x:re.findall(r'Distance from the Dock is (.*?) holy lights',str(x))[0])
df['blessings']=df['blessings'].apply(lambda x:re.findall(r'King blessed the house with (.*?) blessings',x)[0])
df['Guarding_Tower']=df['Guarding_Tower'].apply(lambda x:re.findall(r'Distance from Guarding Tower is (.*?) holy lights',str(x))[0])
df['dining_rooms']=df['dining_rooms'].apply(lambda x:re.findall(r'There are (.*?) dining',str(x))[0])
df['Location']=df['Location'].apply(lambda x:re.findall(r'Location of the house is : (.*?)$',str(x))[0])
df['farm']=df['farm'].apply(lambda x:re.findall(r'There is (.*?) land',str(x))[0])
df['bedroom']=df['bedroom'].apply(lambda x:re.findall(r'There are (.*?) bedrooms\n',str(x))[0])
df['bathrooms']=df['bathrooms'].apply(lambda x:re.findall(r'There are (.*?) bathrooms\n',str(x))[0])
df['Knight_house']=df['Knight_house'].apply(lambda x:re.findall(r'Distance from Knight\'s house is (.*?) holy',str(x))[0])
df['builder']=df['builder'].apply(lambda x:re.findall(r'Housing Prices/(.*?).txt',str(x))[0])
df['River']=df['River'].apply(lambda x:re.findall(r'Distance from the River is (.*?) holy lights',str(x))[0])
```
